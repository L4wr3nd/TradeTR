```python
countries = [state for state in data['Country'].unique()]
years = [x for x in range(2013, 2022)]
years_change = ['13/14', '14/15', '15/16', '16/17', '17/18', '18/19', '19/20', '20/21']
dfGDPchange = pd.DataFrame(columns = ['Country', '13/14', '14/15', '15/16', '16/17', '17/18', '18/19', '19/20', '20/21'])
```

```python
pdata = data

keylist = ['Year', 'Country', 'Total'] + calendar
dict_country = {}
for i in keylist:
    dict_country[i] = None

for state in countries:
    for y in years:
        if state not in pdata[pdata['Year'] == y]['Country'].unique():
            dict_country['Year'] = y
            dict_country['Country'] = state
            dict_country['Total'] = 0
            for m in calendar:
                dict_country[m] = 0
            #print(dict_country, '\n')
            pdata = pdata.append(dict_country, ignore_index = True)
pdata.tail()
```

```python
pdata = pdata.sort_values(by = ['Year', 'Total'], ascending = False)
pdata[pdata['Country'] == 'Melilla']
```

```python
def country_absence():
    x = 0
    for state in countries:
        for y in years:
            if state not in data[data['Year'] == y]['Country'].unique():
                x += 1
    return x
```

```python
#checking
print(data.shape)
print(pdata.shape)
print(pdata.shape[0] - data.shape[0] == pdata[pdata['Total'] == 0].shape[0])
print(pdata.shape[1] == data.shape[1] == pdata[pdata['Total'] == 0].shape[1])
print(country_absence())
print(pdata[pdata['Total'] == 0].shape)
```

```python
pdata.to_excel(r'Imports_by_Countries_formatting_filled.xls', index = False)
```

```python
countryImpchange = {}
for state in countries:
    listTotal = pdata[(pdata['Country'] == state)][['Year', 'Total']].sort_values(by = ['Year'], ascending = True)
    listTotal = listTotal['Total']
    listcountry = []
    for x in range(1,9):
        if listTotal.iloc[x-1] != 0:
            listcountry.append(100*((listTotal.iloc[x]/listTotal.iloc[x-1]) -1))
        else:
            listcountry.append(None)
    countryImpchange[state] = listcountry
```
